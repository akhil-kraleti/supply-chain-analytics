# Data Cleaning Log

In this notebook we prepare the raw supply chain dataset for SQL and Power BI.

Cleaning steps include:
- Removing irrelevant columns
- Removing redundant columns
- Handling missing values
- Correcting data types
- Validating the cleaned dataset

**STEP-1 REMOVING IRRELEVANT COLUMNS**

```python
'''From the EDA, we noticed that,
    --> Product Description for every record is (NaN).
    --> order Zipcode contains 85% of missing values.
    --> Customer Zipcode have only 3 missing values, but this column has no use in our analysis.
    --> Customer Password, Email are personal information and has nothing to do with our analysis.
    --> Product Status just indicates the availability of the product, so we don't need that data.'''
columns_to_drop = [
    "Product Description",
    "Order Zipcode",
    "Customer Zipcode",
    "Customer Password",
    "Customer Email",
    "Product Status"
]

df.drop(columns=columns_to_drop, inplace=True)
```

```python
df.shape
```

```python
'''From the EDA, we noticed that
   -->Customer Id == Order Customer Id , drop one
   -->Category Id == Product Category Id , drop one
   -->Product Card Id == Order Item Cardprod Id , drop one
   -->Product Price == Order Item Product Price , drop one
   -->Benifit Per Order == Order Profit Per Order , drop one
   -->Sales Per Customer == Order Item Total , drop one'''
duplicate_columns = [
    "Order Customer Id",
    "Product Category Id",
    "Order Item Cardprod Id",
    "Order Item Product Price",
    "Order Profit Per Order",
    "Sales per customer"
]

df.drop(columns=duplicate_columns, inplace=True)
```

```python
df.shape
```

**CHECKING IF ANY OTHER MISSING VALUES ARE LEFT**

```python
missing = df.isnull().sum()

missing[missing > 0]
```

**HANDLING THE REMAINING MISSING VALUES**

```python
df["Customer Lname"] = df["Customer Lname"].fillna("Unknown")
```

```python
df.isnull().sum()
```

**HANDLING WRONG DATA TYPES**

```python
df["order date (DateOrders)"] = pd.to_datetime(df["order date (DateOrders)"])

df["shipping date (DateOrders)"] = pd.to_datetime(df["shipping date (DateOrders)"])
```

```python
df.info()
```

**RENAMING COLUMNS**

```python
df.rename(columns={
    "Type": "OrderType",
    "Days for shipping (real)": "ActualShippingDays",
    "Days for shipment (scheduled)": "ScheduledShippingDays",
    "Benefit per order": "Profit",
    "Delivery Status": "DeliveryStatus",
    "Late_delivery_risk": "LateDeliveryRisk",

    "Category Id": "CategoryID",
    "Category Name": "CategoryName",

    "Customer City": "CustomerCity",
    "Customer Country": "CustomerCountry",
    "Customer Fname": "CustomerFirstName",
    "Customer Id": "CustomerID",
    "Customer Lname": "CustomerLastName",
    "Customer Segment": "CustomerSegment",
    "Customer State": "CustomerState",
    "Customer Street": "CustomerStreet",

    "Department Id": "DepartmentID",
    "Department Name": "DepartmentName",

    "Latitude": "Latitude",
    "Longitude": "Longitude",

    "Market": "Market",

    "Order City": "OrderCity",
    "Order Country": "OrderCountry",
    "order date (DateOrders)": "OrderDate",
    "Order Id": "OrderID",

    "Order Item Discount": "DiscountAmount",
    "Order Item Discount Rate": "DiscountRate",
    "Order Item Id": "OrderItemID",
    "Order Item Profit Ratio": "ProfitRatio",
    "Order Item Quantity": "Quantity",

    "Sales": "GrossSales",
    "Order Item Total": "NetSales",

    "Order Region": "OrderRegion",
    "Order State": "OrderState",
    "Order Status": "OrderStatus",

    "Product Card Id": "ProductID",
    "Product Image": "ProductImage",
    "Product Name": "ProductName",
    "Product Price": "ProductPrice",

    "shipping date (DateOrders)": "ShippingDate",
    "Shipping Mode": "ShippingMode"

}, inplace=True)
```

**FINAL VALIDATION**

```python
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

print("\nMissing Values")
print(df.isnull().sum().sum())

print("\nDuplicate Rows")
print(df.duplicated().sum())
```

**SAVING THE CLEANED DATASET**

```python
df.to_csv(
    "../data/cleaned/supply_chain_cleaned.csv",
    index=False
)
```
