**DATA VISUALIZATION**

```python
#Delivery Status Plot
import matplotlib.pyplot as plt

delivery_counts = df["Delivery Status"].value_counts()

plt.figure(figsize=(5,3))
delivery_counts.plot(kind="bar")
plt.title("Delivery Status Distribution")
plt.xlabel("Delivery Status")
plt.ylabel("Number of Orders")
plt.show()
```

```python
#Shipping Mode Plot
shipping_counts = df["Shipping Mode"].value_counts()

plt.figure(figsize=(5,3))
shipping_counts.plot(kind="bar")
plt.title("Shipping Mode Distribution")
plt.xlabel("Shipping Mode")
plt.ylabel("Orders")
plt.show()
```

```python
#Customer Segment Plot
segment_counts = df["Customer Segment"].value_counts()

plt.figure(figsize=(5,3))
segment_counts.plot(kind="bar")
plt.title("Customer Segment")
plt.xlabel("Segment")
plt.ylabel("Orders")
plt.show()
```

```python
#Top 10 Categories 
top_categories = df["Category Name"].value_counts().head(10)

plt.figure(figsize=(10,5))
top_categories.plot(kind="bar")
plt.title("Top 10 Product Categories")
plt.xlabel("Category")
plt.ylabel("Orders")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Standard Class Delivery Status
standard_class_orders = df[df["Shipping Mode"]=="Standard Class"]
Standard_class_delivery_status = standard_class_orders["Delivery Status"].value_counts()
Standard_class_delivery_status.plot(kind="bar")
plt.figure(figsize=(5,3))
plt.show()
```
